```python
import cv2
from matplotlib import pyplot as plt

#to read image from disk,we use 
# cv2.imread function,in below method.
im = cv2.imread("gojo.jpg")
#conter bgr2rgb
color = cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
plt.imshow(color)
plt.title('Image')
plt.show()
```
