# RGB to CMY Conversion


```python
import cv2
print("Original Image")
image = cv2.imread("/content/download.jpeg")
from google.colab.patches import cv2_imshow
cv2_imshow(image)
print("\n\n\nRGB Image")
R,G,B = cv2.split(image)
print("R Complacent")
cv2_imshow(R)

print("G Complacent")
cv2_imshow(G)

print("B Complacent")
cv2_imshow(B)

C = 1-R
M = 1-G
Y = 1-B

print("\n\n\nCMY image")
print("C Complacent")
cv2_imshow(C)

print("M Complacent")
cv2_imshow(M)

print("Y Complacent")
cv2_imshow(Y)
```
